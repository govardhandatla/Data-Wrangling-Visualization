# We Rate Dogs retweets 




## Goal 


Sharing tweets is one of the main keys for growing a twitter account.

Therefore, we use the data at our disposal to get a better understing on what influence sharing, as time (hours, days), kind of dogs (breeds, stages) and rating, in order to give some insight to the marketing team.

## Retweet and time


### Hours and retweets 

NB: the time on the graph is Paris Time GMT.

![Retweets and hours](https://raw.githubusercontent.com/pierreconreaux/wrangle_and_analyze_data/master/Data/hours_and_retweets.png)

We can see that some hours have more retweets, as noon, 4pm (16) or 5 pm (17).

Conversely, 3am, 4am and 1pm underperforms in terms of retweets.


### Days of the week and retweets 

NB: 0 = Monday, 1 = Tuesday, ..., 6 = Sunday.

![Retweets and days](https://raw.githubusercontent.com/pierreconreaux/wrangle_and_analyze_data/master/Data/days_and_retweets.png)

We observe that Tuesday and Friday have a better performance than Saturday and Sunday.


## Retweet and kind of dogs


### Retweet and dog breeds
